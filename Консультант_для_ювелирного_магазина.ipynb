```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS, Chroma
from langchain.docstore.document import Document

# Преобразуем документы в формат LangChain
docs = [Document(page_content=d["content"], metadata={"title": d["title"]}) for d in documents]

# Используем модель эмбеддингов
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Создаём две базы
faiss_db = FAISS.from_documents(docs, embeddings)
chroma_db = Chroma.from_documents(docs, embeddings, collection_name="jewelry_docs")
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.chains import RetrievalQA

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.2)
```
